```python
import numpy as np
from math import comb
import random
import matplotlib.pyplot as plt
```

```python
# Comb is the built in Choose function

total = 0

for i in range(0,12):
  value = 0.5 ** i * comb(i + 12, 2*i)
  total = total + value

total = total * 0.5**12

print(total)
```

```python
total = 0

for i in range(12, 24):
  value = 0.5 ** i * comb(i, 2 * i - 24)
  total = total + value

print(total)
```

```python
moves = [1,2]

reach25 = [0,0] # Yes, no

current_path = 0

for i in range(0,1000000):
  while (current_path < 25):
    current_path = current_path + random.choice(moves)
  if current_path == 25:
    reach25[0] +=1
  else:
    reach25[1] +=1
  current_path = 0

print(reach25)
```

```python
def Prob(n):
  moves = [1,2]

  results = [0,0]

  current_path = 0

  for i in range(0,10000):
    while (current_path < n):
      current_path = current_path + random.choice(moves)
    if current_path == n:
      results[0] +=1
    else:
      results[1] +=1
    current_path = 0

  return results
```

```python
x = 50

n = np.linspace(1,x, x)
wins = np.zeros(x)
for i in range(0,x):
  wins[i] = Prob(i)[0]/10000
```

```python
plt.plot(n, wins)
plt.show()
```
